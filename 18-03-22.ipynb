```python
#Importar librerias pandas 
#Importar libreria numpy as np
#Importar libreria statistics as stat
#importar libreria matplotlib.pyplot as plt
#importar modulo sqlite3

import pandas as pd
import numpy as np
import statistics as stat
import matplotlib.pyplot as plt
import sqlite3
```

```python
# Crea un objeto de conexión a la base de datos SQLite

con = sqlite3.connect('c19.db')

#Usa read_sql_query de pandas para extraer el resultado
# de la consulta a un DataFrame

df = pd.read_sql_query("SELECT * from c19dashboard_uk__national_cases_by_age_gender", con)

# Verifica que el resultado de la consulta SQL está
# almacenado en el DataFrame
print(df.head(20))

# No te olvides de cerrar la conexión
con.close()
```
