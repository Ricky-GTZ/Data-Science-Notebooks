Regresin Lineal Simple con Python
El paquete statsmodel ayuda para hacer la regresin lineal

```python
import pandas as pd
import numpy as np
```

```python
# Leemos dataset
data = pd.read_csv("Advertising.csv")
```

```python
# Explorar el dataset
data.head()
```

```python
# para la regresin lineal 
import statsmodels.formula.api as smf
```

```python
# Relacionamos la variable independiente TV con la variable dependiente Sales
lm = smf.ols(formula="Sales~TV", data = data).fit()
```

```python
lm.params

# Nota el valor de prediccin sera Sales = 7.032594 + 0.047537 * TV 
```

```python
# Determinar el valor del coeficiente de predictibilidad
lm.pvalues
```

```python
lm.summary()
```

```python
# Error de precitibilidad
sales_pred = lm.predict(pd.DataFrame(data["TV"]))
sales_pred
```

```python
#Graficar la regresin lineal
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
data.plot(kind = "scatter", x = "TV", y="Sales")
plt.plot(pd.DataFrame(data["TV"]), sales_pred, c = "red", linewidth= 3)
```
