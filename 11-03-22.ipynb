```python
#Tomar columns
Fs=pd.read_csv("Fwine.csv") #Concatenacion de archivos

c=Fs['free sulfur dioxide']
```

```python
c.head()
```

```python
#Medidas de tendencia central
quartiles=[ ]
Media=np.mean(c)
Mediana=np.median(c)
Moda=stat.mode(c)
quartiles=c
q1=np.median(quartiles[:10])
q2=Media
q3=np.median(quartiles[10:])
IQR=q3-q1
rango=np.percentile(c,100,interpolation='midpoint')  
desviacion_standar=np.std(c)
Varianza=np.var(c)

```

```python
print(Media)
print(Mediana)
print(Moda)
print(q1)
print(q2)
print(q3)
print(IQR)
print(rango)
print(desviacion_standar)
print(Varianza)

```

```python
#Graficar Histogramas
plt.figure(figsize=(15, 10))
histograma=c
plt.hist(histograma, bins=150)
plt.title(label="Histogram free sulfure")
plt.xlabel("wine")
plt.ylabel("Free sulfure dioxide")
```

```python
#Graficar caja de bigote
plt.figure(figsize=(15, 10))

plt.boxplot(c)
plt.title(label="Histogram free sulfure")
plt.xlabel("wine")
plt.ylabel("Free sulfure dioxide")
```
